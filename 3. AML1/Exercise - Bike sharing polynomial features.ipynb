Fit a linear regression and compare its performance to the median baseline using the mean absolute error (MAE) measure.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split
```

```python
# Load data
data_df= pd.read_csv('data/bike-sharing.csv')
data_df.head()
```

```python
# Encode categorical variables
data_df = pd.get_dummies(data_df, columns=['yr', 'workingday', 'holiday', 'weekday', 'season', 'weathersit'])
```

```python
# Split into train/test sets
x= data_df.drop('casual', axis=1).values
y= data_df.casual.values
x_tr, x_te, y_tr, y_te = train_test_split(x, y, train_size=.7, test_size=.3, random_state=0)
```

```python
x_te.shape
```

```python
import numpy as np

# Mean absolute error (MAE)
def MAE(y, y_pred):
    return np.mean(np.abs(y - y_pred))
```

```python
dummy = DummyRegressor(strategy='median')
dummy.fit(x_tr, y_tr)
pred_baseline = dummy.predict(x_te)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_tr, y_tr)
y_pred_lr = lr.predict(x_te)
```

```python
# Median baseline
mae_baseline = MAE(y_te, pred_baseline)

# Linear regression
mae_lr = MAE(y_te, y_pred_lr)

print('MAE baseline: {:.3f}'.format(mae_baseline))
print('MAE linear regression: {:.3f}'.format(mae_lr))
```

## Exercise - Add polynomial features

> **Exercise**: Add the `temp^2` and `temp^3` polynomial features. Then fit and evaluate a linear regression. Plot your model with a scatter plot of temperatures vs. number of users. Feel free to add other features.

```python
# Add polynomial features
x2_tr = np.column_stack((x_tr[:,0],x_tr[:,0]**2, x_tr[:,0]**3, x_tr[:,1:]))
x2_te = np.column_stack((x_te[:,0],x_te[:,0]**2, x_te[:,0]**3, x_te[:,1:]))

# Fit a linear regression
lr2 = LinearRegression()
lr2.fit(x2_tr, y_tr)
y2_pred_lr = lr2.predict(x2_te)

mae_lr2 = MAE(y_te, y2_pred_lr)
print('MAE lr with new features: {:.3f}'.format(mae_lr2))
```

```python
# Plot predictions
sns.set()
plt.figure(figsize=(10, 8))
plt.scatter(x_te[:,0], y_te, label='original values', s=20)
plt.scatter(x_te[:,0], y_pred_lr, label='pred LR', s=20)
plt.legend(loc='upper left')
plt.show()
```

## Exercise - Separate sources

In the last exercise, we saw that we can identify two sources in the data.

1. Data points collected during working days
1. Data points collected during non-working days

The goal of this exercise is to create a model for each source using your extended set of features, e.g., the original features plus the `temp^2`, `temp^3` polynomial features.

> **Exercise**: Create a model for each source with the extended set of features, and evaluate the overall performance on the test set using MAE. Plot the two models with a scatter plot of temperatures vs. number of users. Create a final comparison using a bar chart.

```python
# Separate data point
tr_indWKDAY = x2_tr[:,8]==1
te_indWKDAY = x2_te[:,8]==1

x_tr_wkday = x2_tr[tr_indWKDAY]
y_tr_wkday = y_tr[tr_indWKDAY]

x_te_wkday = x2_te[te_indWKDAY]
y_te_wkday = y_te[te_indWKDAY]
```

```python
x_tr_NOTwkday = x2_tr[~tr_indWKDAY]
y_tr_NOTwkday = y_tr[~tr_indWKDAY]

x_te_NOTwkday = x2_te[~te_indWKDAY]
y_te_NOTwkday = y_te[~te_indWKDAY]
```

```python
# Fit a linear regression for working days (wd)
# and one for non-working days (nwd)
lr_wd = LinearRegression()
lr_wd.fit(x_tr_wkday, y_tr_wkday)
y2_pred_wd = lr_wd.predict(x_te_wkday)

lr_nwd = LinearRegression()
lr_nwd.fit(x_tr_NOTwkday, y_tr_NOTwkday)
y2_pred_nwd = lr_nwd.predict(x_te_NOTwkday)

# Compute overall performance with MAE
mae_wdnwd = MAE(np.append(y2_pred_wd, y2_pred_nwd), np.append(y_te_wkday, y_te_NOTwkday))
print('MAE two sources: {:.3f}'.format(mae_wdnwd))
```

```python
# Plot predictions
plt.figure(figsize=(10, 8))
plt.scatter(x_te_wkday[:,0], y_te_wkday, label='original values', s=20)
plt.scatter(x_te_wkday[:,0], y2_pred_wd, label='pred LR - polyfeats', s=20)
plt.legend(loc='upper left')
plt.show()
```

```python
# Plot predictions
plt.figure(figsize=(10, 8))
plt.scatter(x_te_NOTwkday[:,0], y_te_NOTwkday, label='original values', s=20)
plt.scatter(x_te_NOTwkday[:,0], y2_pred_nwd, label='pred LR - polyfeats', s=20)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x_te[:,0], y_te, label='original values', s=20)
plt.scatter(x_te_NOTwkday[:,0], y2_pred_nwd, label='pred LR NWD', s=20)
plt.scatter(x_te_wkday[:,0], y2_pred_wd, label='pred LR WD', color= 'orange', s=20)
plt.legend(loc='upper left')
plt.show()
```
