```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.dummy import DummyClassifier
```

```python
# Load the numpy .npz file
with np.load('cifar4-train.npz', allow_pickle=False) as data:
    cifar_data = dict(data.items())

X = cifar_data['pixels'].astype(np.float32)
y = cifar_data['labels']
```

```python
X = X/255
```

```python
X[:4, :]
```

```python
cifar_classes = cifar_data['names']
```

### Print the class distribution, i.e., the proportion of images in each category.

```python
df = pd.DataFrame({'data':y})

print(df['data'].value_counts())
```

```python
plt.bar([1, 2, 3, 4], df['data'].value_counts())
plt.xticks([1, 2, 3, 4], cifar_classes[::-1])
plt.show()

### the 4 classes of pictures are equally distributed with 1250 occurences of each. 
```

### What is the accuracy of the "most-frequent" baseline?

```python
### The most frequent dummy gives us the frequency of the msot frequent class, hence 25%. 
### The dummy classifier is not much need, but is given for future reference. 

dummy = DummyClassifier(strategy='most_frequent')

dummy.fit(X, y)

accuracy = dummy.score(X, y)
print('Accuracy: {:.2f}'.format(accuracy))
```

### Plot a few images from the data set, e.g., ten images from each category.

```python
np.random.seed(654321)

for k in range(4):
    fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(8, 3))
    res = np.random.choice([idx for idx,i in enumerate(y) if i==k], size=10, replace=False)
    plt.suptitle(cifar_classes[k]+'s')
    
    for a, axis in enumerate(axes.flatten()): 
        axis.imshow(X[res[a]].reshape(32, 32, 3), interpolation="nearest")
        axis.get_xaxis().set_visible(False) # disable x-axis
        axis.get_yaxis().set_visible(False)

plt.show()
```

### Create a scree plot of the OverFeat features.

```python
overfeat = cifar_data['overfeat']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Create Standard scaler
scaler = StandardScaler()

# Rescale data
overfeat_rescaled = scaler.fit_transform(overfeat)

# Create PCA transformer
pca = PCA()

# Apply PCA to rescaled data
overfeat_pca_rescaled = pca.fit_transform(overfeat_rescaled)
```

```python
pve = pca.explained_variance_ratio_
```

```python
pve_cumsum = np.cumsum(pve)
np.where(pve_cumsum > 0.95)
```

```python
xcor = np.arange(1, 409) 
plt.xticks(xcor)

# Add cumulative sum
plt.plot(xcor, pve_cumsum[:408])

# Add labels
plt.xlabel('principal component')
plt.ylabel('Proportion of variance explained')
plt.xticks(range(1, 409, 81))
plt.title('Scree plot (cumulative sum)')
plt.xlabel('Number of components')
plt.show()
```

```python
print('The PVE of the first 2 components is {:.2f}%'.format(pve_cumsum[1]*100))
```

```python
for percentile in [0.8, 0.9, 0.95, 0.99]:
    print('{:.0f}% of the variance is explained by the first {} components.'.format(percentile * 100,
                                                                                    (pve_cumsum < percentile).sum() + 1))
```

```python
### 2d plot of the first two components on the first 500 data points

overfeat_pca_rescaled_sub = overfeat_pca_rescaled[:500]

for kind in [0, 1, 2, 3]:
    # Wine samples of this type
    idx = (y[:500] == kind)

    # Plot their components
    plt.scatter(
        overfeat_pca_rescaled_sub[idx, 0], overfeat_pca_rescaled_sub[idx, 1],
        label='{}'.format(cifar_classes[kind]+'s')
    )

# Labels and legend
plt.legend()
plt.xlabel('1st component')
plt.ylabel('2nd component')
plt.show()
```

The points overlap a lot; nonetheless, we can observe that the overlap is segregated in two groups: trucks/cars and airplanes/ships. 

This makes sense as features might include common features, eg. wheels, roads, window shields for the former, and sky/sea (blue), body color usually white, and longiline body shape for the latter two. 

```python
### PCAing the matrix of pixels. 

# Create Standard scaler
scaler = StandardScaler()

# Rescale data
pixels_rescaled = scaler.fit_transform(X)

# Create PCA transformer
pca_pix = PCA(n_components=2)

# Apply PCA to rescaled data
pixels_pca_rescaled = pca_pix.fit_transform(pixels_rescaled)
```

```python
### 2d plot of the first two components on the first 500 data points

pixel_pca_rescaled_sub = pixels_pca_rescaled[:500]

for kind in [0, 1, 2, 3]:
    # Wine samples of this type
    idx = (y[:500] == kind)

    # Plot their components
    plt.scatter(
        pixel_pca_rescaled_sub[idx, 0], pixel_pca_rescaled_sub[idx, 1],
        label='{}'.format(cifar_classes[kind]+'s')
    )

# Labels and legend
plt.legend()
plt.xlabel('1st component')
plt.ylabel('2nd component')
plt.show()
```

Contrary to the previous plot, all categories are mixed and the plot is not giving us much insight, even though the PVE of the 2 components is superior in this case. 

```python
print('PVE of the first 2 components for the matrix of image pixels is {:.0f}%'.format(
    pca_pix.explained_variance_ratio_.cumsum()[1] *100))
```

### Preparing the data for further analysis

```python
from sklearn.model_selection import train_test_split


# Split data into train/test sets
X_tr, X_te, y_tr, y_te = train_test_split(
    overfeat, y, train_size=4000, test_size=1000, random_state=0, stratify = y)
```
