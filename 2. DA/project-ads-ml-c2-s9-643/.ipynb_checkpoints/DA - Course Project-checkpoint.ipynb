## B. 3 Interesting Questions

### 1. What are the foods that have the highest ratios of each macronutrient?  

```python
dataSub = data[data.pnns_groups_2 != 'unknown']
```

```python
dataSubMeans = dataSub.groupby('pnns_groups_2')['energy_100g','carbohydrates_100g', 
                                                'proteins_100g', 'fat_100g'].mean()
```

```python
dataSubMeans['sumMacro'] = dataSubMeans.iloc[:, 1:].sum(axis=1)
```

```python
for i in dataSubMeans.columns[1:4]:
    col_name = i.replace('100g', '') + 'ratio'
    dataSubMeans[col_name] = dataSubMeans[i]/dataSubMeans.sumMacro
```

```python
for i in dataSubMeans.columns[5:]:
    dataSubMeans.loc[:, i].sort_values().plot.bar()
    plt.title(i)
    plt.show()
```

### 2. Is there a realtionship between the macronutrient ratio and the total Energy?

```python
for i in dataSubMeans.columns[5:]:
    #sns.lmplot(x=i, y='energy_100g', data=dataSubMeans)
    sns.jointplot(x=i, y='energy_100g', data=dataSubMeans, kind="reg");
```

Fat is positively correlated to energy, while the carbohydrates to energy exhibits a moderate negative relationship. The ratio of proteins is not correlated to energy. 

### 3. What is the macronutrient breakdown across different countries? 

```python
### We need to expand the countries tags in the original dataframe
data['countries_tagsSplit'] = data.countries_tags.str.split(',')
data2 = data.copy()
```

```python
###https://stackoverflow.com/questions/38203352/expand-pandas-dataframe-column-into-multiple-rows
dataSplit = (data2.apply(lambda x: pd.Series(x.countries_tagsSplit), axis=1).stack()
             .reset_index(level=1, drop=1)
             .to_frame('countries')
             .join(data2)
            )
```

```python
dataSplit.head()
```

```python
dataCountry = dataSplit.groupby('countries')['energy_100g','carbohydrates_100g', 
                                                'proteins_100g', 'fat_100g'].mean()
dataCountry.head()
```

```python
dataCountry['sumMacro'] = dataCountry.iloc[:, 1:].sum(axis=1)
for i in dataCountry.columns[1:4]:
    col_name = i.replace('100g', '') + 'ratio'
    dataCountry[col_name] = dataCountry[i]/dataCountry.sumMacro
```

```python
dataCountry.head()
```

```python
### cleaning missing values
dataCountry.dropna(how='any', inplace=True)
```

```python
## countries tags are crowd-sourced level
## we could have done something like "get only countries with at least 5 observations" 
## will do if requested. 
dataCountry.iloc[:, 5:].sort_values(['carbohydrates_ratio']).plot.barh(stacked=True, figsize=(8, 25))
plt.title('Macronutrients Ratio per Country')
plt.legend(loc='upper left')
plt.show()
```

## E. Build a database

```python
## create a connection to a sqlite3 database
db = sqlite3.connect('database.sqlite')
```

```python
def run_query(query):
    return pd.read_sql_query(query,db) 
```

```python
### restrict your data to 1000 entries and 5 columns of your choice 
### create one or multiple tables, at least one of the tables should have a PRIMARY KEY
query= "CREATE TABLE Products (ind INTEGER PRIMARY KEY, created_datetime TEXT, product_name TEXT, brands_tags TEXT, pnns_groups_2 TEXT, energy_100g REAL);"
cursor = db.cursor()
cursor.execute(query)
```

```python
### fill the database with your data
data['ind'] = data.index
data.loc[:1000, ['ind',
                  'created_datetime', 
                  'product_name',
                  'brands_tags',
                  'pnns_groups_2',
                  'energy_100g']].to_sql(name='Products', con=db, if_exists='append', index= False, chunksize=200) 
```

```python
db.commit()
```

```python
### run at least one query to demonstrate that it works correctly
cursor.execute("SELECT product_name, pnns_groups_2 FROM Products WHERE Products.energy_100g > 3000;")
results = cursor.fetchall()
print(results)
```

## F. Finding correlations

```python
data[['nutrition-score-fr_100g', 'energy_100g', 'fat_100g', 'saturated-fat_100g', 
      'carbohydrates_100g', 'sugars_100g', 'fiber_100g',
      'proteins_100g', 'salt_100g', 'sodium_100g']].corr().iloc[1:, :1].sort_values('nutrition-score-fr_100g', 
                                                                                   ascending=False)
```

Energy, saturated fat, fat and sugars are the most positively correlated to the nutrition score, as one could expect, considering that the highest the nutrition score, the least healthy it is. Fibers are negatively correlatd to the nutritional score, which is also consistent with the expected effect.
