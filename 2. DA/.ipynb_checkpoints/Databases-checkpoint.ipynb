```python
import pandas as pd
import sqlite3
```

```python
db = sqlite3.connect('database.sqlite')
```

```python
def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
 run_query("SELECT tbl_name FROM sqlite_master;")
```

```python
 run_query("SELECT * FROM trip LIMIT 10;")
```

```python
 run_query("SELECT start_station_name, end_station_name, duration FROM trip LIMIT 5;")
```

```python
query = '''
SELECT start_station_name, end_station_name, duration FROM trip 
WHERE start_station_name = 'Paseo de San Antonio' 
LIMIT 5;
'''
run_query(query)
```

```python
query='''
SELECT start_station_name, end_station_name, duration FROM trip 
WHERE duration<100 
LIMIT 5
'''
run_query(query)
```

```python
query='''
SELECT start_station_name, end_station_name, duration FROM trip 
WHERE (duration<100) AND (start_station_name = 'Paseo de San Antonio')
LIMIT 5;
'''
run_query(query)
```

```python
query='''
SELECT start_station_name, end_station_name, duration FROM trip 
WHERE start_station_name = 'Paseo de San Antonio' 
ORDER BY duration DESC 
LIMIT 5;
'''
run_query(query)
```

```python
run_query("SELECT AVG(duration) AS 'Average Duration' FROM trip;")
```

```python
query='''
SELECT subscription_type, AVG(duration) AS 'Average Duration'  FROM trip 
GROUP BY subscription_type;
'''
run_query(query)
```

```python
query = '''
SELECT station.city AS "City", COUNT(*) AS "Count"
FROM trip 
INNER JOIN station
ON trip.start_station_name = station.name
GROUP BY station.city;
'''
run_query(query)
```

```python
query = '''
SELECT start.city AS "Start city", end.city AS "End city", trip.duration
FROM trip
INNER JOIN station AS start
ON trip.start_station_name = start.name
INNER JOIN station AS end
ON trip.end_station_name = end.name
WHERE start.city != end.city
ORDER BY duration DESC
LIMIT 1;
'''
run_query(query)
```

```python
query = '''
SELECT sub.*
FROM (SELECT * FROM trip
WHERE duration < 100) sub
WHERE start_station_name='Paseo de San Antonio';
'''
run_query(query)
```

```python
###1) What was the duration of the longest trip?
query = '''
SELECT duration FROM trip
ORDER BY duration DESC
LIMIT 1;
'''
run_query(query)
```

```python
###2) How many trips were ended at the station '2nd at South Park '?
query = '''
SELECT COUNT(*) FROM trip
WHERE end_station_name = "2nd at South Park";
'''
run_query(query)
```

```python
###3) How many trips had a duration between 1000 and 2000?
query = '''
SELECT COUNT(*) FROM trip
WHERE duration BETWEEN 1000 AND 2000;
'''
run_query(query)
```

```python
###4) Which bike was used for the most trips?
query = '''
SELECT bike_id AS 'Bike', COUNT(*) AS 'Number of Trips' FROM trip
GROUP BY bike_id
ORDER BY COUNT(*) DESC
LIMIT 1;
'''
run_query(query)
```

```python
###4) Which bike was used for the most trips?
query = '''
SELECT sub.start_station_name AS 'Station', COUNT(*) AS 'Number of Round Trips' 
FROM (SELECT * FROM trip
WHERE end_station_name == start_station_name) sub
GROUP BY start_station_name
ORDER BY COUNT(*) DESC
LIMIT 5;
'''
run_query(query)
```

```python
query = '''
SELECT submitted_via AS 'Method', COUNT(*) AS 'Number', 
(SELECT COUNT(*) FROM data) as 'Total' FROM data
GROUP BY submitted_via
ORDER BY COUNT(*) DESC;
'''
run_query(query)
```

```python
query = '''
SELECT *,
(CAST (q1.Number AS double precision)/CAST (q1.Total AS double precision))*100 AS 'Percent'
FROM (SELECT submitted_via AS 'Method', COUNT(*) AS 'Number', 
    (SELECT COUNT(*) FROM data) as 'Total' 
FROM data
GROUP BY submitted_via
ORDER BY COUNT(*) DESC) q1;
'''
run_query(query)
```
