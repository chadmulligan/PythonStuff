```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
X = np.arange(-15,16, .4)
W = np.sin(X)/X
Y = 2*np.sin(X)/X
Z = 3*np.sin(X)/X
```

```python
lines1, = plt.plot(X, W, color = "blue", linestyle = '--', label = "sin(x)/x")
lines2, = plt.plot(X, Y, color = 'green', label = "2*sin(x)/x")
lines3, = plt.plot(X, Z, color = 'red', marker = 'o', label = "3*sin(x)/x")

plt.legend()
plt.show()
```

```python
countries = ['USA', 'GB', 'China', 'Russia', 'Germany', 'France']
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 17, 10])
golds = np.array([46, 27, 26, 19, 17])
total = np.array([121,67,70,55,42])
```

```python
x_cor = np.arange(5)
```

```python
plt.bar(x_cor, total, color='green', edgecolor='black')
plt.xticks(x_cor, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.title("2016 Olympics Medals: Top Countries")
plt.bar(x_cor, bronzes, color='#CD853F', label='bronze medals')
plt.bar(x_cor, silvers, color='silver', bottom=bronzes, label='silver medals')
plt.bar(x_cor, golds, color='gold', bottom=silvers+bronzes, label='gold medals')
plt.legend(loc="upper right")
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'], 
columns=iris['feature_names'])
iris_df['species'] = iris['target']
iris_df.head()
```

```python
species = ['Iris setosa', 'Iris versicolor', 'Iris virginica']
colors = ['red', 'orange', 'green']
```

```python
iris_df.loc[iris_df.species == 0, 'species'] = species[0]
iris_df.loc[iris_df.species == 1, 'species'] = species[1]
iris_df.loc[iris_df.species == 2, 'species'] = species[2]
```

```python
setosa = iris_df.loc[iris_df.species=="Iris setosa", :]
versicolor = iris_df.loc[iris_df.species=="Iris versicolor", :]
virginica = iris_df.loc[iris_df.species=="Iris virginica", :]
```

```python
setosa.species.unique()[0]
```

```python
plt.scatter(setosa['sepal length (cm)'], setosa['petal length (cm)'], color = 'red', label = setosa.species.unique()[0])
plt.scatter(versicolor['sepal length (cm)'], versicolor['petal length (cm)'], color = 'orange', label = versicolor.species.unique()[0])
plt.scatter(virginica['sepal length (cm)'], virginica['petal length (cm)'], color = 'green', label = virginica.species.unique()[0])
plt.legend(loc = 4)
```

```python
A = np.array([[3,4],[2,3]])
print(np.dot(A,np.linalg.inv(A)))
np.linalg.matrix_rank(A)
```

```python
b=np.array([[1],[2]])
```

```python
x=np.linalg.solve(A,b)
print(x)
```

```python
np.dot(A,x) == b
```
